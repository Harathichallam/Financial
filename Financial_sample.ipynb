```python
df1=df.drop('Date',axis=1)
```

```python
df1['Discount Band']=df1['Discount Band'].fillna(df1['Discount Band'].mode()[0])
```

```python
df1.isnull().sum()
```

```python
df1.duplicated().value_counts()
```

```python
num1=df1.select_dtypes(['int64','float64']).columns
cat1=df1.select_dtypes('object').columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in num1:
  sns.distplot(df1[i])
  plt.show()
```

```python
for i in num1:
  sns.boxplot(df1[i])
  plt.show()
```

```python
import numpy as np
```

```python
for i in num1:
  q3,q1=df1[i].quantile([0.75,0.25])
  iqr=q3-q1
  upper=q3+1.5*iqr
  lower=q1-1.5*iqr
  df1[i]=np.where(df1[i]>upper,upper,np.where(df1[i]<lower,lower,df1[i]))
```

```python
for i in num1:
  sns.boxplot(df1[i])
  plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
for i in cat1:
  df1[i]=encoder.fit_transform(df1[i])
```

```python
y = df1['Profit']
x = df1.drop(['Profit'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
```

```python
random=RandomForestRegressor(n_estimators=300,
    max_depth=8,
    min_samples_split=8,
    min_samples_leaf=8,
    max_features='sqrt',

    random_state=42)
random.fit(x_train,y_train)
```

```python
random.score(x_train,y_train)
```

```python
random.score(x_test,y_test)
```
